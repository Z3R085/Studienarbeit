```python
# Importieren der ben√∂tigten Bibliotheken
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Laden der CSV-Datei
file_path = 'data_neu.csv'  # Datei im selben Verzeichnis
data = pd.read_csv(file_path)

# Konvertierung des 'messzeitpunkt' in datetime
data['messzeitpunkt'] = pd.to_datetime(data['messzeitpunkt'])
```

```python
# Erstellung des Diagramms
plt.figure(figsize=(10, 6))
# Entfernen des marker-Parameters, um nur die Linie zu zeichnen
plt.plot(data['messzeitpunkt'], data['feuchtigkeitswert'], linestyle='-', color='blue')

# Formatierung des Diagramms
plt.title('Feuchtigkeitsverlauf')
plt.xlabel('Wochentag + Uhrzeit')
plt.ylabel('Feuchtigkeitswert in %')
plt.xticks(rotation=45)

# Anpassen des DateFormatter, um Wochentag und Uhrzeit anzuzeigen
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%a %H:%M'))
plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=1))

plt.grid(True)

# Anzeigen des Diagramms
plt.tight_layout()
plt.show()
```

```python
# Erstellung des Diagramms
plt.figure(figsize=(10, 6))
# Zeichnen des Graphen f√ºr die Temperaturwerte, ohne einzelne Punkte darzustellen
plt.plot(data['messzeitpunkt'], data['temperatur'], linestyle='-', color='red')

# Formatierung des Diagramms
plt.title('Temperaturverlauf')
plt.xlabel('Wochentag + Uhrzeit')
plt.ylabel('Temperatur in ¬∞C')
plt.xticks(rotation=45)

# Anpassen des DateFormatter, um Wochentag und Uhrzeit anzuzeigen
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%a %H:%M'))
plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=1))

plt.grid(True)

# Anzeigen des Diagramms
plt.tight_layout()
plt.show()
```
