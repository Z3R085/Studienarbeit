```python
# Importieren der ben√∂tigten Bibliotheken
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Laden der CSV-Datei
file_path = 'data.csv'  # Datei im selben Verzeichnis
data = pd.read_csv(file_path)

# Konvertierung des 'messzeitpunkt' in datetime
data['messzeitpunkt'] = pd.to_datetime(data['messzeitpunkt'])
```

```python
# Erstellung des Diagramms
plt.figure(figsize=(10, 6))

# Zeichnen der Linie ohne Marker
plt.plot(data['messzeitpunkt'], data['feuchtigkeitswert'], linestyle='-', color='blue')

# Formatierung des Diagramms
plt.title('Feuchtigkeitsverlauf')
plt.xlabel('Wochentag + Datum')
plt.ylabel('Feuchtigkeitswert in %')
plt.xticks(rotation=45)

# Anpassen des DateFormatter, um Wochentag und Datum anzuzeigen
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%a %d.%m.'))
# Anpassen des Locators, um nur einen Tick pro Tag zu setzen
plt.gca().xaxis.set_major_locator(mdates.DayLocator())

plt.grid(True)

# Anzeigen des Diagramms
plt.tight_layout()
plt.show()
```

```python


n = 5
data['moving_average'] = data['feuchtigkeitswert'].rolling(window=n).mean()

# Erstellung des Diagramms
plt.figure(figsize=(10, 6))

# Zeichnen der urspr√ºnglichen Linie
#plt.plot(data['messzeitpunkt'], data['feuchtigkeitswert'], linestyle='-', color='blue', label='Original')

# Zeichnen der gleitenden Durchschnittslinie
plt.plot(data['messzeitpunkt'], data['moving_average'], linestyle='-', color='red', label='Trend (Moving Average)')

# Formatierung des Diagramms
plt.title('Feuchtigkeitsverlauf mit Trendlinie')
plt.xlabel('Wochentag + Datum')
plt.ylabel('Feuchtigkeitswert in %')
plt.xticks(rotation=45)
plt.legend()

# Anpassen des DateFormatter und Locators
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%a %d.%m.'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Erstellung des Diagramms
plt.figure(figsize=(10, 6))

# Zeichnen des Graphen f√ºr die Temperaturwerte, ohne einzelne Punkte darzustellen
plt.plot(data['messzeitpunkt'], data['temperatur'], linestyle='-', color='red')

# Formatierung des Diagramms
plt.title('Temperaturverlauf')
plt.xlabel('Wochentag + Datum')
plt.ylabel('Temperatur in ¬∞C')
plt.xticks(rotation=45)

# Anpassen des DateFormatter, um Wochentag und Datum anzuzeigen
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%a %d.%m.'))
# Anpassen des Locators, um nur einen Tick pro Tag zu setzen
plt.gca().xaxis.set_major_locator(mdates.DayLocator())

plt.grid(True)

# Anzeigen des Diagramms
plt.tight_layout()
plt.show()
```

```python
# Durchschnittstemperatur berechnen und auf zwei Nachkommastellen runden
durchschnittstemperatur = round(data['temperatur'].mean(), 2)
print(f'Durchschnittstemperatur: {durchschnittstemperatur}¬∞C')

# Median der Temperatur berechnen und auf zwei Nachkommastellen runden
median_temperatur = round(data['temperatur'].median(), 2)
print(f'Median der Temperatur: {median_temperatur}¬∞C')

# Durchschnittsfeuchtigkeit berechnen und auf zwei Nachkommastellen runden
durchschnittsfeuchtigkeit = round(data['feuchtigkeitswert'].mean(), 2)
print(f'Durchschnittsfeuchtigkeit: {durchschnittsfeuchtigkeit}%')

# Median der Feuchtigkeit berechnen und auf zwei Nachkommastellen runden
median_feuchtigkeit = round(data['feuchtigkeitswert'].median(), 2)
print(f'Median der Feuchtigkeit: {median_feuchtigkeit}%')

# Feuchtigkeitsschwankungen berechnen (Standardabweichung) und auf zwei Nachkommastellen runden
feuchtigkeitsschwankungen = round(data['feuchtigkeitswert'].std(), 2)
print(f'Feuchtigkeitsschwankungen (Standardabweichung): {feuchtigkeitsschwankungen}')

# Temperaturschwankungen berechnen (Spannweite) und auf zwei Nachkommastellen runden
temperaturschwankungen = round(data['temperatur'].max() - data['temperatur'].min(), 2)
print(f'Temperaturschwankungen (Spannweite): {temperaturschwankungen}¬∞C')
```
